```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import string
import nltk
from gensim.models import Word2Vec
import pandas as pd
import numpy as np
```

```python
nltk.download('punkt')
nltk.download('stopwords')
```

```python
df = pd.read_csv('../data/enc_tok_nopunct_lemm.csv')
df.head()
```

```python
model = Word2Vec(df[['word_tokens']], vector_size=200, window=5, min_count=1, workers=16)

model.save("../embeddings/word2vec_model.model")
```

```python
def get_sentence_embedding(sentence, model):
    word_vectors = [model.wv[word] for word in sentence if word in model.wv]
    if not word_vectors:
        return None
    return sum(word_vectors) / len(word_vectors)
```

```python
df['w2v_embeddings'] = df['word_tokens'].apply(lambda x: get_sentence_embedding(x, model))
```

```python
df[[str(i) for i in range(model.vector_size)]] = pd.DataFrame(df['w2v_embeddings'].tolist())
```
